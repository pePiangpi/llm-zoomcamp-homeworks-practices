```python
from sqlitesearch import TextSearchIndex

sqlite_index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    db_path="faq.db"
)
```

```python
sqlite_index.count()
```

```python
results = sqlite_index.search("Can I still join the course after it started?", num_results=1)
[(doc["question"], doc['answer']) for doc in results]
```
